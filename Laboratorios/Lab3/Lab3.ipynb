**1. Retraso de primer orden vs. cadena de retrasos**

- Retraso de primer orden:

Representa procesos donde la salida depende de un único parámetro de tiempo característico. Se asume la probabilidad de que un stock no depende del tiempo que un individuo ha permanecido en él. La salida es proporcional al stock dividido en el tiempo de retraso.

- Cadena de retrasos:
  
Son varios retrasos de primer ordn conectados en serie para modelar procesos multietapa. Permitiendo simular distribuciones de tiempos de permanencia realistas y suaviza el flujo, eliminando la propiedad de sin memoria.


El retraso de primer orden es más simple pero menos realista para procesos biológicos. En contrastte, las cade3nas de retrasos permiten una mejor aproximacion a la variabilidad  y distribuciones de tiempos de permanencia.


**2. Fuerza de infección (λ) vs. número básico de reproducción (R₀)**

- Fuerza de infección (λ):

 Es la tasa instantánea con la que los suceptibles se infectan. 
 $$
 \lambda = \Beta \frac{I}{N}
 $$



- Número básico de reproducción (R₀):

Representa el número promedio de casos secundarios generados por un caso primario en una población suseptible.
Es un parámetro fijo del agente y contexto inicial. $R_0$ determina el umbral de inmunidad de grupo y la dificultad de control.

- λ es una tasa dinámica, depende del estado de la epidemia $\frac{I}{N}$.

- R₀ es un potencial intrínseco de transmisión; λ puede cambiar, R₀ es fijo si el contexto no cambia.



**3. ¿Por qué el sarampión requiere >95% de cobertura de vacunación y el ébola ~50%?**
Esto se explica por la diferencia en R₀ entre ambos virus.

- Sarampión: 
    - R₀ = 12-18, transmisibilidad alta por vía aérea.
    - Umbral de inmunidad de grupo: $1-\frac{1}{R_0}$, necesita inmunidad de más del 95% de la población para controlar la epidemia.

- Ébola:
    - R₀ = 2-3, transmisibilidad por vía directa.
    - Umbral de inmunidad de grupo: $1-\frac{1}{R_0}$, necesita inmunidad de más del 50% de la población para controlar la epidemia.

Mientrás mayor sea el $R_0$, mayor población inmune para la prevención de brotes.



# Ejercicio 2 – Retroalimentación retardada y oscilaciones

## Escenario 1

Un país con cobertura de vacunación contra la COVID-19 del 60 % observa un aumento de \(R_t\) de 0,8 a 1,5 tras la aparición de una nueva variante.

1. **Factores que explican el incremento de \(R_t\)**  
   - **Escape inmunitario**: la variante presenta mutaciones en la proteína Spike que reducen la neutralización por anticuerpos inducidos por la vacuna.  
   - **Relajación de medidas de control**: disminución del uso de mascarillas y del distanciamiento, lo que aumenta la tasa de contacto.

2. **Intervención económica**  
   - **Incentivo monetario para dosis de refuerzo**  
     - Ejemplo: vale de transporte o bono al recibir la dosis adicional.  
     - *Objetivo:* reducir barreras de costo y tiempo, elevando la cobertura de refuerzo.

3. **Intervención no económica**  
   - **Campaña informativa de riesgo**  
     - Mensajes en medios y redes sociales sobre la mayor transmisibilidad de la variante.  
     - *Objetivo:* aumentar la percepción de riesgo y fomentar el uso de mascarillas y el distanciamiento.

## Escenario 2

En una región con vacunación contra el VPH del 40 % y asistencia escolar del 80 %, se compara la eficacia de un mandato de vacunación escolar frente a clínicas móviles gratuitas. El \(R_0\) estimado del VPH es 4.

1. **Umbral de inmunidad de grupo**  
   $$
   \text{Cobertura crítica} \;=\; 1 \;-\; \frac{1}{R_0}
   \;=\; 1 \;-\; \frac{1}{4}
   \;=\; 0{,}75 \; (75\%)
   $$

2. **Comparación de estrategias**  
   - **Mandato escolar**  
     - Cobertura potencial: hasta el 80 % (vacunación obligatoria para matricularse).  
     - Elimina la opción de “free-riding” y supera rápidamente el 75 % necesario.  
   - **Clínicas móviles gratuitas**  
     - Cobertura depende de la decisión individual; difícilmente supera el 75 %.  
     - Persiste un dilema de coordinación: algunos optarán por no vacunarse confiando en que otros sí lo hagan.

3. **Justificación desde teoría de juegos**  
   - El mandato escolar establece un equilibrio de Nash en el que vacunarse es la mejor estrategia individual (si no, no se puede matricular).  
   - Las clínicas móviles mantienen un dilema voluntario, limitando la cobertura alcanzable.
