**1. Retraso de primer orden vs. cadena de retrasos**

- Retraso de primer orden:

Representa procesos donde la salida depende de un único parámetro de tiempo característico. Se asume la probabilidad de que un stock no depende del tiempo que un individuo ha permanecido en él. La salida es proporcional al stock dividido en el tiempo de retraso.

- Cadena de retrasos:
  
Son varios retrasos de primer ordn conectados en serie para modelar procesos multietapa. Permitiendo simular distribuciones de tiempos de permanencia realistas y suaviza el flujo, eliminando la propiedad de sin memoria.


El retraso de primer orden es más simple pero menos realista para procesos biológicos. En contrastte, las cade3nas de retrasos permiten una mejor aproximacion a la variabilidad  y distribuciones de tiempos de permanencia.


**2. Fuerza de infección (λ) vs. número básico de reproducción (R₀)**

- Fuerza de infección (λ):

 Es la tasa instantánea con la que los suceptibles se infectan. 
 $$
 \lambda = \Beta \frac{I}{N}
 $$



- Número básico de reproducción (R₀):

Representa el número promedio de casos secundarios generados por un caso primario en una población suseptible.
Es un parámetro fijo del agente y contexto inicial. $R_0$ determina el umbral de inmunidad de grupo y la dificultad de control.

- λ es una tasa dinámica, depende del estado de la epidemia $\frac{I}{N}$.

- R₀ es un potencial intrínseco de transmisión; λ puede cambiar, R₀ es fijo si el contexto no cambia.



**3. ¿Por qué el sarampión requiere >95% de cobertura de vacunación y el ébola ~50%?**
Esto se explica por la diferencia en R₀ entre ambos virus.

- Sarampión: 
    - R₀ = 12-18, transmisibilidad alta por vía aérea.
    - Umbral de inmunidad de grupo: $1-\frac{1}{R_0}$, necesita inmunidad de más del 95% de la población para controlar la epidemia.

- Ébola:
    - R₀ = 2-3, transmisibilidad por vía directa.
    - Umbral de inmunidad de grupo: $1-\frac{1}{R_0}$, necesita inmunidad de más del 50% de la población para controlar la epidemia.

Mientrás mayor sea el $R_0$, mayor población inmune para la prevención de brotes.

**4. Defina y compare los mecanismos de: La lotería de vacunas de Ohio vs Las exigencias de ingreso escolar en Francia**

- La lotería de vacunas de Ohio:
    - Es una intervenciòn conductual en donde las personas vacunadas participan en loterías para obtener un premio.

    - El mecanismo usado es un incentivo a la poblaciòn que apela a la motivación intrínseca de las personas.El deseo de ganar un premio, no una obligación, reduciendo la fricciòn o resistencia a la vacunación. Usa tecnicas conductuales como la gamificaciòn y como resultado se espera un aumento en la tasa de vacunación en individuos con baja motivación o indesición. Depende del valor del premio y la publicidad del sorteo.
- Las exigencias de ingreso escolar en Francia:
    - Es una política pública donde la vacunación es un requisito legal y obligatorio para acceder a servicios como la educación preescolar o primaria.

    - Aumenta el costo de no vacunarse al imponer barreras institucionales: quien no cumple con el calendario de vacunación no puede matricularse en la escuela. Es un cambio en la estructura de incentivos y sanciones, no solo en la motivación personal.

    - Asegura una cobertura masiva de vacunación, pero puede generar resistencia social y legal.

        | **Naturaleza**           | Incentivo positivo/voluntario      | Restricción obligatoria/legal              |
        | **Mecanismo**            | Motivación extrínseca por premio   | Penalización estructural por no cumplir    |

## Escenario 2

En una región con vacunación contra el VPH del 40 % y asistencia escolar del 80 %, se compara la eficacia de un mandato de vacunación escolar frente a clínicas móviles gratuitas. El \(R_0\) estimado del VPH es 4.

1. **Umbral de inmunidad de grupo**  
   $$
   \text{Cobertura crítica} \;=\; 1 \;-\; \frac{1}{R_0}
   \;=\; 1 \;-\; \frac{1}{4}
   \;=\; 0{,}75 \; (75\%)
   $$

2. **Comparación de estrategias**  
   - **Mandato escolar**  
     - Cobertura potencial: hasta el 80 % (vacunación obligatoria para matricularse).  
     - Elimina la opción de “free-riding” y supera rápidamente el 75 % necesario.  
   - **Clínicas móviles gratuitas**  
     - Cobertura depende de la decisión individual; difícilmente supera el 75 %.  
     - Persiste un dilema de coordinación: algunos optarán por no vacunarse confiando en que otros sí lo hagan.

3. **Justificación desde teoría de juegos**  
   - El mandato escolar establece un equilibrio de Nash en el que vacunarse es la mejor estrategia individual (si no, no se puede matricular).  
   - Las clínicas móviles mantienen un dilema voluntario, limitando la cobertura alcanzable.


# **Ejercicio 3**

**Promt utilizado:**

Necesito que me ayudes a implementar en Python una simulación de un modelo SIR con scipy.integrate.odeint para tres escenarios: (a) sin intervenciones, (b) intervención tipo 'lotería' donde la tasa de vacunación aumenta en 0.02/día cuando los casos acumulados superen los 100, y (c) intervención tipo 'mandato' donde se refuerza la vacunación de inmediato al 50% de la población cuando Rₜ > 1. Los parámetros base son β=0.35, γ=0.1, infectados iniciales=10, población=10,000. Quiero el código con comentarios claros en español, gráficos comparativos y el análisis final del número de infectados. Por favor no uses librerías como Mesa o PySD.

**Porque este promt me funcionó:**

- Especifiqué claramente el modelo matemático (SIR) y la librería (odeint) que debía usarse.
- Detallé los escenarios y condiciones de intervención para que la IA no asumiera supuestos erróneos.
- Incluí parámetros numéricos concretos para que el código fuera reproducible sin pedir valores extra.
- Aclaré las debidas restricciones para no usar modelos preconstruidos, con el fin de cumplir con las reglas del laboratorio.
- Solicité comentarios claros en español y en primera persona, alineados con lo que exige la entrega para poder entender que esta sucediendo (no los incluí en el laboratorio si no que fue más para entender que estaba sucediendo por dentro).


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

N = 10000          # Población total
I0 = 10            # Infectados iniciales
R0 = 0             # Recuperados iniciales
S0 = N - I0 - R0   # Susceptibles iniciales
beta = 0.35        # Tasa de transmisión
gamma = 0.1        # Tasa de recuperación
dias = 160         # Días de simulación

def sir(y, t, beta, gamma, politica):
    S, I, R = y
    vacunacion = 0

    if politica == "loteria" and I > 100:
        vacunacion = 0.02 * S  # Vacunación diaria adicional

    if politica == "mandato":
        Rt = (beta / gamma) * (S / N)  # Número reproductivo efectivo
        if Rt > 1:
            vacunacion = 0.5 * S  # Vacunar inmediatamente al 50% de los susceptibles

    dSdt = -beta * S * I / N - vacunacion
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I + vacunacion
    return dSdt, dIdt, dRdt

def simular(politica):
    t = np.linspace(0, dias, dias)
    y0 = S0, I0, R0
    sol = odeint(sir, y0, t, args=(beta, gamma, politica))
    S, I, R = sol.T
    return t, S, I, R

t_a, S_a, I_a, R_a = simular("baseline")
t_b, S_b, I_b, R_b = simular("loteria")
t_c, S_c, I_c, R_c = simular("mandato")

plt.figure(figsize=(10,6))
plt.plot(t_a, I_a, label="Sin intervención")
plt.plot(t_b, I_b, label="Lotería (↑ vacunación tras 100 casos)")
plt.plot(t_c, I_c, label="Mandato (50% vacunación si Rₜ>1)")
plt.xlabel("Días")
plt.ylabel("Número de infectados")
plt.title("Simulación SIR con diferentes intervenciones")
plt.legend()
plt.grid(True)
plt.show()

infectados_final_a = R_a[-1]
infectados_final_b = R_b[-1]
infectados_final_c = R_c[-1]

print("Comparación de infectados finales:")
print(f"Sin intervención: {infectados_final_a:.0f}")
print(f"Lotería: {infectados_final_b:.0f}")
print(f"Mandato: {infectados_final_c:.0f}")
```

Al analizar la solución que obtuve, puedo observar que el comportamiento del modelo sigue la lógica esperada para cada política. En el escenario sin intervención, la curva de infectados crece rápidamente, alcanzando un pico alto y disminuyendo después, lo que refleja un brote no controlado. El número final de infectados (9660) es casi toda la población susceptible, lo que confirma que, sin medidas, la enfermedad se propaga ampliamente.

En el escenario de “lotería”, la intervención de aumentar la vacunación cuando los casos superan 100 logra reducir el pico máximo de contagios y aplanar ligeramente la curva. Sin embargo, la diferencia en el número final de infectados con respecto a la línea base es nula, lo que indica que esta medida es poco efectiva si se implementa tarde o con una intensidad limitada. La forma de la curva demuestra que el virus aún tiene suficiente margen para propagarse antes de que la vacunación haga un impacto significativo.

Finalmente, el escenario de “mandato” muestra un cambio drástico en la dinámica. La vacunación inmediata del 50 % de los susceptibles cuando $R_t$ supera 1 genera una reducción pronunciada en la curva de infectados: el pico prácticamente desaparece y el número final de infectados cae a 7302. Esto evidencia que la intervención temprana y masiva es mucho más eficaz para cortar la cadena de transmisión y prevenir contagios. Visualmente, la curva verde se mantiene plana, lo que refleja control casi total de la epidemia después de la intervención.
